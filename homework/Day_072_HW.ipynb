# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

x = np.arange(-10,10,0.01)
```

```python
def ReLU(x):
    return abs(x) * (x > 0)

plt.plot(x, ReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('ReLU Function')
plt.show()
```

```python
def dReLU(x):
    return (1 * (x > 0))

plt.plot(x, dReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('dReLU Function')
plt.show()
```
