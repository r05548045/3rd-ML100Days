```python
print("-----AMT_GOODS_PRICE")
print("Mean: ", app_train['AMT_GOODS_PRICE'].mean())
print("Standard Deviation: ", app_train['AMT_GOODS_PRICE'].std())

print("-----AMT_CREDIT")
print("Mean: ", app_train['AMT_CREDIT'].mean())
print("Standard Deviation: ", app_train['AMT_CREDIT'].std())

print("-----AMT_ANNUITY")
print("Mean: ", app_train['AMT_ANNUITY'].mean())
print("Standard Deviation: ", app_train['AMT_ANNUITY'].std())

fig, axes = plt.subplots(1, 3)
app_train['AMT_GOODS_PRICE'].hist(ax=axes[0], grid=False, bins=25)
app_train['AMT_CREDIT'].hist(ax=axes[1], grid=False, bins=25)
app_train['AMT_ANNUITY'].hist(ax=axes[2], grid=False, bins=25)
axes[0].set_title('AMT_GOODS_PRICE')
axes[1].set_title('AMT_CREDIT')
axes[2].set_title('AMT_ANNUITY')
```
