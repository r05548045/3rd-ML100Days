```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MeanAbsoluteError(y1, y2):
    mae = sum(abs(y1 - y2)) / len(y1)
    return mae

def MeanSquaredError(y1, y2):
    mse = sum(y1 - y2) / len(y1)
    return mse
```

```python
w = 2
b = 0.25
x_line = np.linspace(0, 100, 101)
y = (x_line + np.random.randn(101) * 4) * w + b

plt.plot(x_line, y, 'b.', label = 'data points')
plt.title("Data points")
plt.legend(loc = 2)
plt.show()
```

```python
y_ = x_line * w + b
plt.plot(x_line, y, 'b.', label = 'data')
plt.plot(x_line, y_, 'r-', label = 'prediction')
plt.title("Data points with prediction")
plt.legend(loc = 2)
plt.show()
```

```python
MSE = MeanSquaredError(y, y_)
MAE = MeanAbsoluteError(y, y_)
print("The Mean squared error is %.2f" % (MSE))
print("The Mean absolute error is %.2f" % (MAE))
```
